- 编制程序，实现数据零均值化
- 编制程序，实现数据最小-最大规范化

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 生成数据
x = np.arange(0, 1, 0.001)
y = stats.beta.pdf(x, 4, 2) # scipy.stats #生成指定分布, beta：beta分布; pdf：概率密度函数
# 求均值
mean_x = np.mean(x)
mean_y = np.mean(y)
# 零均值化
x1 = x - mean_x
y1 = y - mean_y

fig = plt.figure()
# 2行2列 第一个
plt.subplot(2, 2, 1)
plt.title("original data")
plt.plot(x, y, "k-", linewidth = 2)
plt.scatter(mean_x, mean_y, color = 'k', marker = 'o')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 2.12)
# 2行2列 第二个
plt.subplot(2, 2, 2)
plt.title("zero-mean data", loc = 'center')
plt.plot(x1, y1, 'b-', linewidth = 1)
plt.scatter(np.mean(x1), np.mean(y1), color = 'b', marker = 'x')
plt.grid(True)
plt.xlim(-0.5, 0.5)
plt.ylim(-1, 1.12)
# 2行一列 第二行
plt.subplot(2,1,2)
plt.title("data comparison", loc = 'center', y = -0.3)
plt.plot(x, y, 'k-', linewidth = 2)
plt.scatter(mean_x, mean_y, color = 'k', marker = 'o')
plt.plot(x1, y1, 'b-', linewidth = 1)
plt.scatter(np.mean(x1), np.mean(y1), color = 'b', marker = 'x')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import scipy.stats as sts
import matplotlib as mpl
import matplotlib.pyplot as plt

# 绘图工具参数配置
mpl.rcParams["axes.grid"] = "True"
mpl.rcParams["grid.color"] = "darkgrey"
mpl.rcParams["lines.linewidth"] = 1
mpl.rcParams["axes.prop_cycle"] = mpl.cycler(color = ["grey"])

x1 = np.random.normal(50, 15, 100)
z1 = (x1 - np.min(x1)) / (max(x1) - min(x1))

fig = plt.figure()
plt.subplot(221)
plt.plot(x1)
plt.plot(z1, c = 'y')
plt.subplot(222)
plt.hist(x1)
plt.subplot(223)
plt.plot(z1, c = 'y')
plt.subplot(224)
plt.hist(z1)
plt.show()

print(sts.describe(x1))
print(sts.describe(z1))
```
